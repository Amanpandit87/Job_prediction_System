### salary_range columns correction

```python
def salry(text):
    text = text.split("-")[0]
    return text
```

```python
df["sal1"] = df["salary_range"].apply(salry)
```

```python
df.head()
```

```python
def salry(text):
    text = text.split("-")[1]
    return text
```

```python
df["sal2"] = df["salary_range"].apply(salry)
```

```python
df.head()
```

```python
df = df.drop(columns = ["salary_range"])
```

```python
df.head()
```

```python
def clean_salary(val):
    if isinstance(val, str):
        val = val.lower().strip()
        if 'k' in val:
            val = val.replace('k', '')
            return int(float(val) * 1000)
        elif 'lac' in val:
            val = val.replace('lac', '')
            return int(float(val) * 100000)
        elif val.isnumeric():
            return int(val)
        else:
            return None
    elif isinstance(val, (int, float)):
        return int(val)
    else:
        return None
```

```python
df['salary_cleaned_1'] = df['sal1'].apply(clean_salary)
```

```python
df.head()
```

```python
df['salary_cleaned_2'] = df['sal2'].apply(clean_salary)
```

```python
df.head()
```

```python
df.drop(columns = ["sal1","sal2"] ,inplace = True)
```

```python
df.head()
```

```python
df["salary"] = (df["salary_cleaned_1"] + df["salary_cleaned_2"])/2
```

```python
df.head()
```

```python
df.drop(columns = ["salary_cleaned_1","salary_cleaned_2"],inplace = True)
```

```python
df.head()
```

### description columns cleaning

```python
df["description"][0]
```

```python
import re
import string
def text_cleaning(text):
    text = text.lower()
    text = re.sub("\[.*?\]"," ",text)
    text = re.sub("\\W"," ",text)
    text = re.sub("https?//\S+|www\.\S+"," ",text)
    text = re.sub("<.*?>+"," ",text)
    text = re.sub("[%s]"% re.escape(string.punctuation)," ",text)
    text = re.sub("\n"," ",text)
    text = re.sub("\w*\d\w*"," ",text)
    return text
```

```python
df["description"] = df["description"].apply(text_cleaning)
```

### Removing comma from requirment columns

```python
def comma(text):
    text = text.lower()
    text = re.sub(r",","",text)
    return text
```

```python
df["requirment"] = df["requirment"].apply(comma)
```

```python
x = df.drop(columns = ["name"])
y = df["name"]
```

```python
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
```

```python
y = lab.fit_transform(y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.preprocessing import MinMaxScaler
```

```python
trans = ColumnTransformer(transformers = [("tfidf_req",TfidfVectorizer(),"requirment"),
                                         ("tfidf_des",TfidfVectorizer(),"description"),
                                         ("minmax",MinMaxScaler(),["salary"])],remainder = "passthrough")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
x_train = trans.fit_transform(x_train)
x_test = trans.transform(x_test)
```
